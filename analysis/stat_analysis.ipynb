```python
from itertools import combinations
from scipy.stats import ranksums
import numpy as np 
import pandas as pd 
from scipy.stats import t
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

```python

# Sample data (replace with your own data)
varied = pd.read_csv('varied_eval_AND_AND.csv')['independence']
varied_XOR = pd.read_csv('varied_eval_XOR_AND.csv')['independence']

seq = pd.read_csv('seq_eval_AND_AND.csv')['independence']
seq_XOR = pd.read_csv('seq_eval_XOR_AND.csv')['independence']

simult = pd.read_csv('simult_eval_AND_AND.csv')['independence']
simult_XOR = pd.read_csv('simult_eval_XOR_AND.csv')['independence']

tri = pd.read_csv('tri_obj_eval_AND_AND.csv')['independence']
tri_XOR = pd.read_csv('tri_obj_eval_XOR_AND.csv')['independence']

# random = pd.read_csv('random_materials.csv')['independence']

# groups = [varied, seq, simult,tri ]
# group_names = ['varied', 'seq', 'simult','tri']

groups = [varied, seq, simult,tri]
groups = [varied_XOR, seq_XOR, simult_XOR,tri_XOR]

group_names = ['varied', 'seq', 'simult','tri']

# Perform pairwise comparisons
for group_pair in combinations(zip(groups, group_names), 2):
    group1_data, group1_name = group_pair[0]
    group2_data, group2_name = group_pair[1]
    
    statistic, p_value = mannwhitneyu(group1_data, group2_data)#,alternative='greater')
    p_value*=6 #Bonferroni correction 
    print(f"Comparison between {group1_name} and {group2_name}:")
    print("Wilcoxon rank-sum statistic:", statistic)
    print("p-value:", p_value)
    
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis. There is a significant difference between the groups.")
    else:
        print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
    print()
```

```python
for group_data, group_name in zip(groups, group_names):
    mean = np.mean(group_data)
    n = len(group_data)
    std_err = np.std(group_data, ddof=1) / np.sqrt(n)
    
    # Calculate t-value for 95% CI with n-1 degrees of freedom
    t_value = t.ppf(0.975, df=n-1)
    
    # Calculate margin of error
    margin_of_error = t_value * std_err
    # print(mar)
    # Calculate CI
    lower_ci = mean - margin_of_error
    upper_ci = mean + margin_of_error
    
    print(f"{group_name}:")
    print("Mean:", mean)
    print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_ci, upper_ci))
    print(f'{margin_of_error}')
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataframes
rand_cases = pd.read_csv('correct_cases_rand.csv', dtype={'input_case_pair': str})
seq_cases = pd.read_csv('correct_cases_seq.csv', dtype={'input_case_pair': str})
simult_cases = pd.read_csv('correct_cases_simult.csv', dtype={'input_case_pair': str})

# Reformat 'input_case_pair' column to have all four digits
rand_cases['input_case_pair'] = rand_cases['input_case_pair'].apply(lambda x: x.zfill(4))
seq_cases['input_case_pair'] = seq_cases['input_case_pair'].apply(lambda x: x.zfill(4))
simult_cases['input_case_pair'] = simult_cases['input_case_pair'].apply(lambda x: x.zfill(4))

# Create subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 9))

# Plot bar graph for rand_cases
ax1.bar(rand_cases['input_case_pair'], rand_cases['mats_correct'], color='purple', alpha=0.5)
ax1.set_title('Random Evaluation', fontsize=16)
# ax1.set_xlabel('Input Case Pair')
ax1.set_ylabel('Mats Behaving Correct',fontsize=16)
ax1.set_xticklabels(rand_cases['input_case_pair'], rotation=90,fontsize=16)

# Plot bar graph for seq_cases
for case_pair in seq_cases['input_case_pair']:
    alpha = 1.0 if case_pair in ['0001', '0010', '0011', '1000', '0100', '1100'] else 0.3
    ax2.bar(case_pair, seq_cases[seq_cases['input_case_pair'] == case_pair]['mats_correct'].values[0],
            color='blue', alpha=alpha)
ax2.set_title('Sequential Evaluation', fontsize=16)
# ax2.set_xlabel('Input Case Pair')
ax2.set_ylabel('Mats. Behaving Correct',fontsize=16)
ax2.set_xticklabels(seq_cases['input_case_pair'], rotation=90, fontsize=16)

# Plot bar graph for simult_cases
for case_pair in simult_cases['input_case_pair']:
    alpha = 1.0 if case_pair in ['1010', '0101', '1111'] else 0.3
    ax3.bar(case_pair, simult_cases[simult_cases['input_case_pair'] == case_pair]['mats_correct'].values[0],
            color='green', alpha=alpha)
ax3.set_title('Simultaneous Evaluation', fontsize=16)
ax3.set_xlabel('Input Case Pair',fontsize=16)
ax3.set_ylabel('Mats. Behaving  Correct',fontsize=16)

ax3.set_xticklabels(simult_cases['input_case_pair'], rotation=90,fontsize=16)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
from scipy.stats import chi2_contingency

rand_table = pd.read_csv('correct_cases_rand.csv', dtype={'input_case_pair': str})['mats_correct']
seq_table = pd.read_csv('correct_cases_seq.csv', dtype={'input_case_pair': str})['mats_correct']
simult_table = pd.read_csv('correct_cases_simult.csv', dtype={'input_case_pair': str})['mats_correct']

contingency_table = np.array([rand_table, seq_table, simult_table])

# Perform the chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output the results
print("Chi-squared statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:", expected)
```
